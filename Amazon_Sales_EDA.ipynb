**HANDLING MISSING VALUES AND DROPPING IRRELEVANT COLUMNS**

```python
Amazon_sale_df.drop(['Unnamed: 22'], axis=1, inplace=True)
```

```python
Amazon_sale_df.drop(['fulfilled-by'], axis=1, inplace=True)
```

```python
Amazon_sale_df['Courier Status'].fillna('Unknown', inplace=True)
```

```python
Amazon_sale_df['promotion-ids'].fillna('No Promos', inplace=True)
```

```python
Amazon_sale_df['currency'].fillna(Amazon_sale_df['currency'].mode()[0], inplace=True)
```

```python
Amazon_sale_df_backup = Amazon_sale_df.copy()
```

```python
Amazon_sale_df[Amazon_sale_df['Amount'].isna()][['Status','Order ID','SKU','Qty','Amount']].head(50)
```

```python
Amazon_sale_df[Amazon_sale_df['Amount'].isna()]['Status'].value_counts(dropna=False)
```

```python
median_value = Amazon_sale_df['Amount'].median(skipna=True)
```

```python
mask_returned = Amazon_sale_df['Status'].str.contains('Returned to Seller', na=False)
print("Returned rows with missing Amount before:", Amazon_sale_df[mask_returned & Amazon_sale_df['Amount'].isna()].shape[0])

Amazon_sale_df.loc[mask_returned & Amazon_sale_df['Amount'].isna(), 'Amount'] = 0

print("Returned rows with missing Amount after:", Amazon_sale_df[mask_returned & Amazon_sale_df['Amount'].isna()].shape[0])
```

```python
mask_pending = Amazon_sale_df['Status'].str.strip().str.lower() == 'pending'
print("Pending rows with missing Amount before:", Amazon_sale_df[mask_pending & Amazon_sale_df['Amount'].isna()].shape[0])

Amazon_sale_df.loc[mask_pending & Amazon_sale_df['Amount'].isna(), 'Amount'] = 0

print("Pending rows with missing Amount after:", Amazon_sale_df[mask_pending & Amazon_sale_df['Amount'].isna()].shape[0])
```

```python
mask_shipped_delivered = Amazon_sale_df['Status'].str.contains('Delivered to Buyer|Shipped', na=False, case=False) \
                        & ~Amazon_sale_df['Status'].str.contains('Returned to Seller', na=False, case=False)

print("Shipped/Delivered rows with missing Amount before:", Amazon_sale_df[mask_shipped_delivered & Amazon_sale_df['Amount'].isna()].shape[0])

Amazon_sale_df.loc[mask_shipped_delivered & Amazon_sale_df['Amount'].isna(), 'Amount'] = median_value

print("Shipped/Delivered rows with missing Amount after:", Amazon_sale_df[mask_shipped_delivered & Amazon_sale_df['Amount'].isna()].shape[0])
```

```python
Amazon_sale_df[Amazon_sale_df['Amount'].isna()][['Status','Order ID','SKU','Qty','Amount']]
```

```python
Amazon_sale_df.loc[
    Amazon_sale_df['Status'].str.strip().str.lower() == 'shipping',
    'Amount'] = Amazon_sale_df.loc[
  Amazon_sale_df['Status'].str.strip().str.lower() == 'shipping',
    'Amount'].fillna(median_value)
```

```python
Amazon_sale_df.dropna(
    subset=['ship-city', 'ship-state', 'ship-postal-code', 'ship-country'],
    inplace=True)
```

```python
Amazon_sale_df['Status'].value_counts()
```

```python
Amazon_sale_df.describe()
```

```python
Amazon_sale_df['Date'] = pd.to_datetime(Amazon_sale_df['Date'], errors='coerce')
```

```python
Amazon_sale_df['Date'].isna().sum()
```

```python
Amazon_sale_df['Year'] = Amazon_sale_df['Date'].dt.year
Amazon_sale_df['Month'] = Amazon_sale_df['Date'].dt.month
Amazon_sale_df['Month_Name'] = Amazon_sale_df['Date'].dt.strftime('%B')
Amazon_sale_df['Day'] = Amazon_sale_df['Date'].dt.day
Amazon_sale_df['Weekday'] = Amazon_sale_df['Date'].dt.strftime('%A')
```

```python
Amazon_sale_df.to_csv('/content/Amazon_Sales_Cleaned.csv', index=False)
```
